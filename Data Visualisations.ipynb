```python
x_coordinates=np.array([1,2,3])*1.5
plt.bar(x_coordinates-0.25,[10,20,15],width=0.5,label="Current Year",tick_label=["Gold","Silver","Diamond"])
plt.bar(x_coordinates+0.25,[20,10,12],width=0.5,label="Next Year")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.title("Prices of Metals")
plt.legend()
plt.show()
```

```python
subjects="Phy","Chem","Math","Eng","Eco","CS"
weightage=[20,10,30,15,25,40]
plt.pie(weightage,labels=subjects,explode=(0,0,0,0.5,0,0),autopct='%1.1f%%',shadow=True)
plt.title("Subjects")
plt.show()
```

# Histogram

```python
xsn=np.random.randn(100)
sigma=5
u=70
X=np.round(xsn*sigma+u)
Y=np.round(xsn*5+40)
print(X)
```

```python
plt.style.use("seaborn")
plt.hist(X,label="Maths")
plt.hist(Y,label="Physics")
plt.xlabel("Marks Range")
plt.ylabel("Prob/freq")
plt.title("Histogram")
plt.legend()
plt.show()
```

# Movie Dataset Visualisation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read a CSV file
df=pd.read_csv("https://raw.githubusercontent.com/prateek27/machine-learning-aug-18/master/class_02/movie_metadata.csv")
```

```python
df.head(n=10)
```

```python
df.columns
```

```python
titles=list(df.get('movie_title'))
```

```python
print(titles)
```

```python
freq_titles={}
```

```python
for title in titles:
    length=len(title)
    
    if freq_titles.get(length) is None:
        freq_titles[length]=1
    else:
        freq_titles[length]+=1
```

```python
print(freq_titles)
```

```python
X=np.array(list(freq_titles.keys()))
Y=np.array(list(freq_titles.values()))
print(X,Y)
```

```python
plt.scatter(X,Y)
plt.xlabel("Length of Movie Title")
plt.ylabel("Freq Count")
plt.title("Movie length visualisation")
plt.show()
```
