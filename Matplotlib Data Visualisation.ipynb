```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Generate Normal Distribution
u=5
sigma=2
# Standard Normal Distribution
#Sigma is basically spread in curve
vals=u+ sigma*np.random.randn(1000)
print(vals.shape)
```

```python
plt.hist(vals,100)
plt.show()
```

```python
x=vals
y=np.zeros(x.shape)

plt.scatter(x,y)
plt.show()
```

# Mutivariate Normal Distribution

```python
# BiVariate
mean=np.array([0.0,0.0])
cov=np.array([[1,0],[0,1]])

dist=np.random.multivariate_normal(mean,cov,500)
print(dist.shape)
```

```python
plt.scatter(dist[:,0],dist[:,1])
plt.show()
```

```python
# BiVariate
mean=np.array([0.0,0.0])
cov=np.array([[1,0.8],[0.8,1]])

mean2=np.array([5.0,6.0])
cov2=np.array([[1.3,0.2],[0.2,1.3]])

dist=np.random.multivariate_normal(mean,cov,500)
dist2=np.random.multivariate_normal(mean2,cov2,500)


print(dist.shape)
print(dist2.shape)
```

```python
plt.scatter(dist[:,0],dist[:,1])

plt.scatter(dist2[:,0],dist2[:,1])

plt.show()
```
