```python
from urllib.request import urlopen
```

```python
ipl_url="https://en.wikipedia.org/wiki/Indian_Premier_League"
```

```python
ipl_data=urlopen(ipl_url)
```

```python
print(type(ipl_data))
```

```python
ipl_html=ipl_data.read()
```

```python
print(ipl_html)
ipl_data.close()
```

# Parsing DATA

```python
from bs4 import BeautifulSoup as soup
```

```python
ipl_soup=soup(ipl_html,'html.parser')
print(ipl_soup)
```

```python
ipl_soup.h1
```

```python
tables=ipl_soup.findAll('table',{'class':'wikitable plainrowheaders'})
print(len(tables))
```

```python
ipl_table=tables[0]
print(ipl_table)
```

# Extracting Usefull Information
1. Remove undesired tags
2. Extract Table header and data

```python
headers=ipl_table.findAll('th')
print(len(headers))
```

```python
column_titles=[ct.text[:-1] for ct in headers[:-13]]
print(column_titles)
```

```python
rows_data=ipl_table.findAll('tr')[2:]
print(len(rows_data))
```

```python
first_row=rows_data[0].findAll('td',{})
for d in first_row:
    print(d.text)
```

```python
table_rows=[]
for row in rows_data:
    current_row=[]
    row_data=row.findAll('td',{})
    for idx,data in enumerate(row_data):
        current_row.append(data.text[:-1])
        
    table_rows.append(current_row)
```

```python
print(table_rows)
```

# Writing and Reading CSV files
### 1. CSV stand for Comma Separated Files

```python
filename='IPL History.csv'
with open (filename,'w') as f:
    # write the header
    head_string=','.join(column_titles)
    head_string+='\n'
    f.write(head_string)
    
    for row in table_rows:
        row_string=""
        row_string=','.join(row)
        #row_string=row_string[:-1]
        row_string+='\n'
        f.write(row_string)
```
