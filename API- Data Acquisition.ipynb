# Image Scraping
### Pass-it-on.com (quotes image)

```python
import bs4
import requests
```

```python
url="https://www.passiton.com/inspirational-quotes?page=2"
```

```python
response=requests.get(url)
```

```python
response.content
```

```python
# HTML Parser
soup=bs4.BeautifulSoup(response.content)
```

```python
div_element=soup.findAll('section')
```

```python
div_element[1]
```

```python
image=div_element[1]
```

```python
image.img.attrs['src']
```

```python
with open('inspiration.png','wb') as file:
    img_url=image.img.attrs['src']
    response=requests.get(img_url)
    
    file.write(response.content)
```

```python
for i, article in enumerate(div_element):
    with open('inspired{}.png'.format(i),'wb') as file:
        img_url=image.img.attrs['src']
        response=requests.get(img_url)
        
        file.write(response.content)
```
